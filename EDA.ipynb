```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mtcars.csv", encoding='latin1')
```

```python
cat_cols = ['cyl', 'vs', 'am', 'gear', 'carb']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
df.dtypes
```

```python
num_cols = ['mpg', 'disp', 'hp', 'drat', 'wt', 'qsec']
df[num_cols].hist(bins=10, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# MPG vs some important predictors
for col in ['wt', 'hp', 'disp', 'qsec']:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=col, y='mpg', hue='cyl')
    plt.title(f"mpg vs {col}")
    plt.show()
```

```python
sns.pairplot(df, vars=['mpg', 'wt', 'hp', 'disp'], hue='cyl')
plt.show()
```

```python
corr = df[num_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
